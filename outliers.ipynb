```python
import pandas as pd

df = pd.read_csv("AB_NYC_2019.csv")
df.head()
```

```python
max_threshold = (df['price']/df['minimum_nights']).quantile(0.99)
max_threshold
```

```python
df2 = df[(df['price']/df['minimum_nights'])>max_threshold]
```

```python
min_threshold = (df['price']/df['minimum_nights']).quantile(0.01)
min_threshold
```

```python
min_threshold_df = df[(df['price']/df['minimum_nights'])<min_threshold]
min_threshold_df
```

```python
df3 = df[((df['price']/df['minimum_nights'])>min_threshold)  & ((df['price']/df['minimum_nights']) < max_threshold)]
```
